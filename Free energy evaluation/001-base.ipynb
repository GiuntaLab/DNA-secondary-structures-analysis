The below code converts a sequences in the .csv file to individual fasta files in the `input_seqs` folder categorized by region

```python
#seqDF = pd.read_csv("./region_refs/master_all_seq.csv")
#seqDF = pd.read_csv("five_hundred_seqs.csv")
seqDF = pd.read_csv("1000rpe1hap2seqs.csv") #make a change here
file_path = "./input_seqs/"
outfile_path = "./output/" #new variable
statusDF = pd.DataFrame(
    {
        "seq_id": [],
        "region": [],
        "file": [],
        "status": [],
    }
)

for index in range(0, len(seqDF)):
    start = seqDF["start"][index]
    end = seqDF["stop"][index]
    chromosome = seqDF["chr"][index]
    region = seqDF["region"][index]
    sequence_id = f"{chromosome}_{region}__{start}_{end}"
    sequences = {sequence_id: seqDF["sequence"][index]}

    #os.system(f"mkdir -p {file_path}/{region}/")
    os.makedirs(f"{file_path}/{region}/", exist_ok=True)
    seq_filename = f"{region}/{sequence_id}.fa"
    
    write_fasta(file_path+seq_filename, sequences)

    #os.system(f"mkdir -p ./output/{region}/")
    os.makedirs(f"{outfile_path}/{region}/", exist_ok=True)
    output_file = f"./output/{region}/{sequence_id.replace('.fa', '_f.out')}"

    new_row = pd.DataFrame(
        {
            "seq_id": [sequence_id],
            "region": [region],
            "file": [seq_filename],
            "status": [0],
        }
    )
    statusDF = pd.concat([statusDF, new_row], ignore_index=True)
```

```python
import multiprocessing

# Function to be executed in parallel by each process
def worker_function(num):
    result = num * num
    print(f"Process {num}: Result = {result}")

if __name__ == "__main__":
    num_processes = 4

    # Create a list of processes
    processes = []

    # Start the specified number of processes
    for i in range(num_processes):
        process = multiprocessing.Process(target=worker_function, args=(i,))
        processes.append(process)
        process.start()

    # Wait for all processes to complete
    for process in processes:
        process.join()

    print("All processes have finished")
```

```python
import re
def get_content(content):
    mfe_pattern = r'\((-?\d+\.\d+)\)'
    fete_pattern = r'\[(-?\d+\.\d+)\]'
    #csmfe_d_pattern = r'{\s*([-+]?\d+\.\d+)\s*d=([-+]?\d+\.\d+)\s*}'
    #fme_pattern = r'(\d+\.\d+e[+-]\d+)'
    ed_pattern = r'ensemble diversity (-?\d+\.\d+)'

    mfe = re.findall(mfe_pattern, content)
    fete = re.findall(fete_pattern, content)
    ed = re.findall(ed_pattern, content)
    #csmfe_d = re.findall(csmfe_d_pattern, content)
    #fme = re.findall(fme_pattern, content)
    
    array = [mfe[0] if mfe else None, fete[0] if fete else None, ed[0] if ed else None]
    #array = [mfe[0],fete[0],ed[0]]
    '''
        csmfe_d[0][0],
        csmfe_d[0][1],
        fme[0],'''
        

    return array
```
